```python
from pathlib import Path
```

```python
def check_path_depth(fp, parent):
    print(fp)
    print(parent)
    depth = 0
    while fp.parent != parent:
        depth += 1
        fp = fp.parent
        print(fp)
    return depth
```

```python
fp = Path('/home/rdcrlzh1s1/miniforge3/LICENSE.txt')
parent= Path('/home/rdcrlzh1s/')
check_path_depth(fp, parent)
```
